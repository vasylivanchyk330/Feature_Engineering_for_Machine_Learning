# COMPLETE CASE ANALYSIS


Complete-case analysis (CCA) consists of discarding observations where values in any of the variables are missing. Complete Case Analysis means literally analysing only those observations for which there is information in all of the variables in the dataset. 
<br>
<br>
__Dtypes__:<br>
- categories variables<br>
- numerical variables.

__Assumptions__:<br>
- missing values completely at random (MCAR).<br>
- no more than 5% of the total dataset contains missing data

__Dataset__:<br>
- House Prices dataset with the shape of (1460, 81)

__Content__:
1. Loading Data and Basic Data Exploration
2. Preparing Data for CCA
3. Difference in Value Percentage before and After CCA
    - Numerical values loss visualisation
    - Categorical values loss calculation

## 1. Loading Data and Basic Data Exploration

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# st column display to all columns
pd.set_option('display.max_columns', None)
```

```python
# importing ds and explore the shape

data = pd.read_csv('../houseprice.csv')
data.shape
```

```python
# find the columns with at least 1 missing observation

cols_with_na = [col for col in data.columns if data[col].isnull().mean() > 0]
cols_with_na
```

```python
# find out whether they are numerical or categorical
data[cols_with_na].dtypes
```

There are both numerical and categorical variables with missing observations.

## 2. Preparing Data for CCA

Steps:
- find out the percentage of observations missing per variable:
- capture na variables
- calculate the percentage of observations with complete cases and if <5%:
- remove observations with na in any column from the main house pricing ds

```python
# find out the percentage of observations missing per variable:

# calculate the percentage of missing
# and
# transform the array into a dataframe
data_na = pd.DataFrame(data[cols_with_na].isnull().mean().reset_index())

# add column names to the dataframe
data_na.columns = ['columns', 'na_percentage']

# order the dataframe according to percentage of na per variable
data_na.sort_values(by='na_percentage', ascending=False, inplace=True)

data_na
```

```python
# capture na variables with no or less than 5% NA

cols_cca = [var for var in data.columns if data[var].isnull().mean() < 0.05]
cols_cca
```

```python
# calculate percentage of observations with complete cases

len(data[cols_cca].dropna()) / len(data)
```

```python
# we have <5% observations to be dropped which is good 

# remove observations with na in any column from the main house pricing ds

data_cca = data[cols_cca].dropna()
data.shape, data_cca.shape
```

## 3. Difference in Value Percentage before and After CCA

### - Categorical values loss calculation

```python
# the function capturing the percentage of observations
# for each category in the original and complete case dataset
# and puts them together in a new dataframe

def categorical_distribution(df, df_cca, variable):
    tmp = pd.concat(
        [
            # percentage of observations per category, original data
            df[variable].value_counts() / len(df),

            # percentage of observations per category, cca data
            df_cca[variable].value_counts() / len(df_cca)
        ],
        axis=1)

    # add column names
    tmp.columns = ['original', 'cca']

    return tmp
```

```python
# run the function in a categorical variable
categorical_distribution(data, data_cca, 'BsmtQual')
```

```python
categorical_distribution(data, data_cca, 'MasVnrType')
```

```python
categorical_distribution(data, data_cca, 'SaleCondition')
```

Conclusion: the distribution of houses in each of the categories, is very similar in the original and complete case dataset, which again, is what is expected if the data is missing completely at random, and the percentage of missing data is small.
