## Engineering Rare Categories

Rare values are categories within a categorical variable that are present only in a small percentage of the observations. There is no rule of thumb to determine how small is a small percentage, but typically, any value below 5 % can be considered rare.

Infrequent labels are so few, that it is hard to derive reliable information from them. But more importantly, infrequent labels tend to appear only on train set or only on the test set which may cause over-fitting or not scoring them respectively.

Therefore, to avoid this behaviour, they are grouped into a new category called 'Rare' or 'Other'.

Highly cardinal variables tend to have many infrequent or rare categories, whereas low cardinal variables, may have only 1 or 2 rare labels.

## 1. First Steps

```python
# load the house price dataset

data = pd.read_csv('../houseprice.csv')
data.head()
```

The identification of rare labels should be done using only the training set, and then propagated to the test set. Rare labels should be identified in the training set only. Also, all categories present in test set, not present in the list of non-rare categories derived from the training set, should be treated as rare values and re-grouped into 'Rare'.

```python
# divide into train and test set

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['SalePrice'], axis=1),
    data.SalePrice, 
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

## 2. Exploring Labels Categories

### - Variables with one dominant category

```python
# explore a few examples in which variables have only a few categories,
# say less than 3

for col in X_train.columns:
    
    # if the variable is categorical
    if X_train[col].dtypes == 'O':
        
        # if the variable has less than 3 categories
        if X_train[col].nunique() < 3: 
            
            # print percentage of observations per category
            print(X_train.groupby(col)[col].count() / len(X_train))
            print()
```

#### Conclusion:
The 3 variables above, Street, Utilities and CentralAir, show one dominating category which accounts for more than 93-99% of the observations. Re-grouping the rare label in this situation does not make any sense. We could determine if these variables are useful with exploratory analysis, or any feature selection algorithm, or drop the variables altogether.

### - Variables with few categories

```python
# say, capture the columns with only 4 different labels

for col in X_train.columns:
    
    # if the variable is categorical
    if X_train[col].dtypes == 'O':
        
        # 4-9 categories
        if X_train[col].nunique() == 4: 
            
            # print percentage of observations per category
            print(X_train.groupby(col)[col].count() / len(X_train))
            print()
```

The variables above have only 4 categories, and in all three cases, there is at least one category that is infrequent, that is, that is present in less than 5% of the observations.

When the variable has only a few categories, then perhaps it makes no sense to re-categorise the rare labels into something else.

For example, the first variable MasVnrType shows only 1 rare label, BrkCmn. Thus, re-categorising it into a new label will leave the variable in the same situation.

The second variable ExterQual, contains 2 rare labels Ex and Fa, group these 2 into a new label called 'Rare'.

The third variable BsmtCond contains 3 rare labels, Fa, Gd and Po, so group these 3 under the new label 'Rare'.

### - Variable with high cardinality

```python
# explore examples in which variables have several categories, 
# say more than 10

multi_cat_cols = []

for col in X_train.columns:
    
    # if variable  is categorical
    if X_train[col].dtypes =='O': 
    
        # and has more than 10 categories
        if X_train[col].nunique() > 10:
            
            # add to the list
            multi_cat_cols.append(col)  
            
            # observation % within each category
            print(X_train.groupby(col)[col].count()/ len(X_train), "\n")
```

Many categories are rare in the 3 categorical variables printed above. Plot them:

```python
for col in ['Neighborhood', 'Exterior1st', 'Exterior2nd']:

    temp_df = pd.Series(X_train[col].value_counts() / len(X_train) )

    # make plot with the above percentages
    fig = temp_df.sort_values(ascending=False).plot.bar()
    fig.set_xlabel(col)

    # add a line at 5 % to flag the threshold for rare categories
    fig.axhline(y=0.05, color='red')
    fig.set_ylabel('Percentage of houses')
    plt.show()
```

## 4. Re-grouping Rare Labels with Pandas

```python
# functions for finding the rare/nonrare labels in a category

def find_rare_labels(df, variable, tolerance):
    
    # lable %, whole variable ==1
    temp = df.groupby([variable])[variable].count() / len(df)
    
    rare = [x for x in temp.loc[temp<=tolerance].index.values]
    
    return rare

def find_non_rare_labels(df, variable, tolerance):
    
    # lable %, whole variable ==1
    temp = df.groupby([variable])[variable].count() / len(df)
    
    non_rare = [x for x in temp.loc[temp>tolerance].index.values]
    
    return non_rare

```

```python
# Neighborhood rare labels
find_rare_labels(X_train, 'Neighborhood', 0.05)
```

```python
# Neighborhood non_rare labels

[x for x in X_train['Neighborhood'].unique(
) if x not in find_rare_labels(X_train, 'Neighborhood', 0.05)]
```

```python
def rare_encoding(X_train, X_test, variable, tolerance):

    X_train = X_train.copy()
    X_test = X_test.copy()

    # find the most frequent category
    non_rare_cat = find_non_rare_labels(X_train, variable, tolerance)

    # re-group rare labels
    X_train[variable] = np.where(X_train[variable].isin(
        non_rare_cat), X_train[variable], 'Rare')
    
    X_test[variable] = np.where(X_test[variable].isin(
        non_rare_cat), X_test[variable], 'Rare')

    return X_train, X_test
```

```python
# loop over the categorical variables
for variable in X_train.columns:
    
    X_train, X_test = rare_encoding(X_train, X_test, variable, 0.05)
```

```python
# demostrate some of encode features

for col in ['Neighborhood', 'Exterior1st', 'Exterior2nd']:

    temp_df = pd.Series( X_train[col].value_counts() / len(X_train) )

    # make plot with the above percentages
    fig = temp_df.sort_values(ascending=False).plot.bar()
    fig.set_xlabel(col)

    # add a line at 5 % to flag the threshold for rare categories
    fig.axhline(y=0.05, color='red')
    fig.set_ylabel('Percentage of houses')
    plt.show()
```
